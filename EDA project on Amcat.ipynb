## treatment of outlier of salary column

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df["Salary"]<(Q1 - 1.5 * IQR)) | (df["Salary"] >(Q3 + 1.5 * IQR))]
outliers.shape
```

```python
df1 = df.drop(outliers.index)
df1.shape
```

### box plot after removing outliers

```python
df1["Salary"].plot(kind='box')
```

### kde plot of salary after outliers removal

```python
df1["Salary"].plot(kind='kde')
```

```python
stats.probplot(df1["Salary"], dist="norm", plot=plt)
plt.title('Probability Plot - Normal Distribution')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

#### it is like as normal distribution after removing outliers from Salary column
